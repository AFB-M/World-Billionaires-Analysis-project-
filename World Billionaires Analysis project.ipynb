```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Billionaire.csv")
df.head()
```

```python
df.shape
```

```python
df = df.dropna()
```

```python
print(df.isnull().sum())
```

```python
df["NetWorth"] = df["NetWorth"].str.strip("$")
df["NetWorth"] = df["NetWorth"].str.strip("B")
df["NetWorth"] = df["NetWorth"].astype(float)
```

```python
df.head()
```

```python
dfn = df.sort_values(by=["NetWorth"], ascending=False).head(10)
plt.figure(figsize=(20, 10))
plt.bar(dfn["Name"],	dfn["NetWorth"])
plt.show()
```

```python
df.info()
```

```python
b = df["Source"].value_counts().head()
index =b.index
print(index)
Sources = b.values
print(Sources)
plt.figure(figsize=(5,5))
plt.pie(Sources,labels=index);
```

```python
b = df["Industry"].value_counts().head(15)
index =b.index
print(index)
Industry = b.values
print(Industry)
plt.figure(figsize=(5,5))
plt.pie(Industry, labels=index);
```

```python
b = df["Country"].value_counts().head(10)
index =b.index
print(index)
Country = b.values
print(Country)
plt.figure(figsize=(5,5))
plt.pie(Country, labels=index);
```
